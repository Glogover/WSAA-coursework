```python
import requests # Importing the requests library to make HTTP requests
import json # Importing the json library to work with JSON data
```

```python
# Define the URL for the API
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"
```

```python
response = requests.get(url, timeout=60) # Making a GET request to the API with a timeout of 60 seconds
response.raise_for_status() # Checking if the request was successful, if not it will raise an HTTP error

data = response.json() # Parsing the response from the API as JSON and storing it in the variable 'data'

with open("cso.json", "w", encoding="utf-8") as f: # Opening a file named 'cso.json' in write mode with UTF-8 encoding
    json.dump(data, f, indent=2, ensure_ascii=False) # Writing the JSON data to the file with an indentation of 2 spaces and ensuring that non-ASCII characters are preserved
```
